```python
import concurrent.futures
import pyparsing as pp
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
REDFIN_URL="https://www.redfin.com{}"
REDFIN_ZIP_URL="/zipcode/{zip_code}"

REDFIN_API_CLASS_DEF=("a",{"class":"downloadLink"})
```

```python
zip_list=[]

def _parse_zip_codes(city_state,lat_tuner,lon_tuner):
        
    city_df=scraper.zip_database[(scraper.zip_database['type']=='STANDARD')
                              &(scraper.zip_database['decommissioned']==0)
                              &((scraper.zip_database['primary_city'].str.contains(city_state[0].capitalize()))
                                |(scraper.zip_database['acceptable_cities'].str.contains(city_state[0].capitalize())))
                              &(scraper.zip_database['state']==(city_state[1].upper()))]
    
    lat_lim,lon_lim=_lat_lon_limits(city_df,lat_tuner,lon_tuner)

    zip_series=city_df[(city_df['longitude']>lon_lim[0])&
                  (city_df['longitude']<lon_lim[1])&
                  (city_df['latitude']>lat_lim[0])&
                  (city_df['latitude']<lat_lim[1])]['zip']
    
    zip_list=list(map(str,zip_series))
    
    return zip_list






def _lat_lon_limits(df,lat_tuner,lon_tuner) -> tuple[tuple,tuple]:
    lon_avg=df['longitude'].mean()
    lat_avg=df['latitude'].mean()

    lon_stdev=df['longitude'].std()
    lat_stdev=df['latitude'].std()

    lon_upper_lim=lon_avg+(lon_tuner*lon_stdev)
    lon_lower_lim=lon_avg-(lon_tuner*lon_stdev)

    lat_upper_lim=lat_avg+(lat_tuner*lat_stdev)
    lat_lower_lim=lat_avg-(lat_tuner*lat_stdev)

    lat_lim=(lat_lower_lim,lat_upper_lim)
    lon_lim=(lon_lower_lim,lon_upper_lim)

    return lat_lim,lon_lim
```

```python
city_states=[('houston','tx')]
```

```python
for city_state in city_states:
    zip_list+=(_parse_zip_codes(city_state,1.5,1.5))
```

```python
def _randomized_UA():
        num_var=np.random.randint(100,1000)
        num_var3=np.random.randint(10,100)
        num_var2=num_var3%10
        num_var4=np.random.randint(1000,10000)
        num_var5=np.random.randint(100,1000)

        user_agent={"User-Agent": f"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/{num_var5}.36 (KHTML, like Gecko) "+
                f"Chrome/51.{num_var2}.2704.{num_var} Safari/537.{num_var3} OPR/38.0.{num_var4}.41"}
        
        return user_agent

def _parse_csv(csv_file:str) -> list[list]:
            
        csv_contents_list=[]

        split_csv_file=csv_file.splitlines()

        for li in split_csv_file:
            csv_contents_list.append(pp.common.comma_separated_list.parseString(li).asList())

        return csv_contents_list



def _threaded_request(func,urls):
    responses=[]

    with concurrent.futures.ThreadPoolExecutor() as executor:
        future_to_url = {executor.submit(func,url):url for url in urls}
        for future in concurrent.futures.as_completed(future_to_url):
            url = future_to_url[future]
            try:
                responses.append(future.result())
            except:
                pass
                
    return responses


def _generate_urls(**kwargs):
    urls=[]
    try:
        for zip in kwargs['zip_codes']:
            urls.append(REDFIN_URL.format(REDFIN_ZIP_URL.format(zip_code=zip)))
    except:
        pass

    try:
        for link in kwargs['api_links']:
            urls.append(REDFIN_URL.format(link))
    except:
        pass
        
    return urls



def _get_soup(url):
    while True:
        header=_randomized_UA()

        req=requests.get(url,headers=header)


        req_text=req.text

        soup=BeautifulSoup(req_text,'html.parser')

        break

    return soup





def _get_API_response(url):
    header=_randomized_UA()

    req=requests.get(url,headers=header)

    req_text=req.text

    parsed_response=_parse_csv(req_text)

    return parsed_response

def _get_API_links(soups:list[BeautifulSoup]):
        api_links=[]
        for soup in soups:
            try:
                target=soup.find(REDFIN_API_CLASS_DEF[0],REDFIN_API_CLASS_DEF[1])['href']
                api_links.append(target)
            except:
                api_links.append(None)
        return api_links
```

```python
api_links=[]
zip_list=['78705'] #Eliminate duplicates


page_urls=_generate_urls(zip_codes=zip_list)

soups=_threaded_request(_get_soup,urls=page_urls)

api_links=_get_API_links(soups)


api_urls=_generate_urls(api_links=api_links)

api_responses=_threaded_request(_get_API_response,urls=api_urls)

temp_df_list=[]

for response in api_responses:
    temp_df=pd.DataFrame(data=response[1:],columns=response[0])
    temp_df_list.append(temp_df)

final_df=pd.concat(temp_df_list,axis=0)
final_df=final_df.apply(lambda row:pd.to_numeric(row,errors='ignore'))
final_df.reset_index(inplace=True)

df=final_df





```
